```python


# Import necessary libraries
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize.treebank import TreebankWordDetokenizer
from nltk.cluster.util import cosine_distance
import numpy as np

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Function to preprocess and clean the text
def preprocess_text(text):
    sentences = sent_tokenize(text)
    words = [word.lower() for word in word_tokenize(text) if word.isalnum()]
    return sentences, words

# Function to calculate sentence scores based on word frequency
def calculate_sentence_scores(sentences, word_frequencies):
    sentence_scores = {}

    for sentence in sentences:
        for word, freq in word_frequencies.items():
            if word in sentence.lower():
                if sentence in sentence_scores:
                    sentence_scores[sentence] += freq
                else:
                    sentence_scores[sentence] = freq

    return sentence_scores

# Function to generate a summary
def generate_summary(text, num_sentences=2):
    sentences, words = preprocess_text(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.lower() not in stop_words]

    # Calculate word frequencies
    word_frequencies = FreqDist(words)

    # Calculate sentence scores
    sentence_scores = calculate_sentence_scores(sentences, word_frequencies)

    # Get the most important sentences
    summary_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:num_sentences]

    # Detokenize the summary sentences
    summary = TreebankWordDetokenizer().detokenize(summary_sentences)

    return summary

# Example usage
complex_text = """
The stock market, a dynamic and intricate financial ecosystem, serves as a conduit for the buying and selling of securities, primarily stocks, representing ownership in publicly traded companies. At its core, the stock market operates on the principles of supply and demand, where investors engage in a perpetual pursuit of value and profitability. Market participants, ranging from individual retail investors to institutional behemoths, interact within a vast electronic network, engaging in transactions that impact asset prices and market indices.
The market's efficiency is underpinned by information dissemination, as participants analyze financial statements, economic indicators, and geopolitical events to forecast market movements. Investors employ various strategies, such as fundamental analysis, which evaluates a company's financial health, and technical analysis, which studies historical price patterns, to make informed decisions. Market fluctuations, influenced by macroeconomic factors, investor sentiment, and global events, contribute to the volatility that defines the stock market's nature.
Central to stock market functionality are exchanges, like the New York Stock Exchange (NYSE) and NASDAQ, which facilitate the buying and selling of securities through regulated platforms. The market's regulatory framework, enforced by entities such as the Securities and Exchange Commission (SEC), aims to maintain fairness, transparency, and investor protection. The concept of market indices, such as the S&P 500 or Dow Jones Industrial Average, encapsulates the performance of specific segments of the market and aids in benchmarking investment returns.
Market participants also engage in various financial instruments, including options and derivatives, to hedge risk or amplify potential returns, adding layers of complexity to the market's intricacies. Overall, the stock market serves as a barometer of economic health, reflecting the collective expectations, assessments, and decisions of a diverse array of investors navigating the complex landscape of global finance..
"""

summary = generate_summary(complex_text)
print("Summary:")
print(summary)
```
